```python
from sklearn import datasets

datos = datasets.load_breast_cancer(as_frame=True)

datos.frame.head()
```

```python
from sklearn.model_selection import train_test_split

x = datos.data
y = datos.target

X_train, X_test, y_train, y_test = train_test_split(x,y, test_size=0.3)
```

```python
from sklearn import svm
from sklearn import metrics


kernels = ['linear', 'rbf', 'sigmoid']
gammas = [1, 0.1, 0.01, 0.001]

for kernel in kernels:
    for gamma in gammas:
      clf = svm.SVC(kernel=kernel, gamma=gamma)
      clf.fit(X_train, y_train)
      prediciones = clf.predict(X_test)
      print(f"kernel= {kernel}, gamma= {gamma} con accuracy:", metrics.accuracy_score(y_test, prediciones))
      print(f"kernel= {kernel}, gamma= {gamma} con precision:", metrics.precision_score(y_test, prediciones))
```

```python
clf = svm.SVC(kernel="linear", gamma=gamma)
clf.fit(X_train, y_train)
prediciones = clf.predict(X_test)
print(f"kernel= {kernel}, gamma= {gamma} con accuracy:", metrics.accuracy_score(y_test, prediciones))
print(f"kernel= {kernel}, gamma= {gamma} con precision:", metrics.precision_score(y_test, prediciones))
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, prediciones))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(y_test, prediciones)
```
