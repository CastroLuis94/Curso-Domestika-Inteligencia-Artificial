```python
import pandas as pd

datos = pd.read_csv("U3_02_Celsius.csv")

print(datos)
```

```python
x, y = datos["celsius"], datos["fahrenheit"]

print(x,y)
```

```python
import tensorflow as tf

modelo = tf.keras.Sequential([
    tf.keras.Input(shape=(1,)),
    tf.keras.layers.Dense(units=1)
])

modelo.compile(optimizer=tf.keras.optimizers.Adam(.15) , loss="mean_squared_error")

entrenamiento = modelo.fit(x, y, epochs=1000)
```

```python
import matplotlib.pyplot as plt


plt.xlabel("Epocas")
plt.ylabel("Error")
plt.title("Error vs epocas")
plt.plot(entrenamiento.history["loss"])
```

```python
import numpy as np
grados_celsius = [35, 40, 45, 50, 55, 60, 65,123]
for i in grados_celsius:
    print(modelo.predict(np.array([i])))
```
